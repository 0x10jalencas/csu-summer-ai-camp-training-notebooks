```python
# Data Preprocessing Functions
print("🔧 DATA PREPROCESSING SETUP")
print("=" * 80)

def preprocess_features(df, feature_columns):
    """
    Preprocess the feature columns for machine learning
    """
    # Create a copy of the data with only our feature columns
    X = df[feature_columns].copy()
    
    print(f"Original feature matrix shape: {X.shape}")
    
    # Check for missing values
    missing_counts = X.isnull().sum()
    missing_columns = missing_counts[missing_counts > 0]
    
    if len(missing_columns) > 0:
        print(f"\n⚠️  Missing values found:")
        for col, count in missing_columns.items():
            print(f"  - {col}: {count} missing values")
        
        # Handle missing values (fill with median for numeric, mode for categorical)
        for col in missing_columns.index:
            if X[col].dtype in ['int64', 'float64']:
                X[col].fillna(X[col].median(), inplace=True)
            else:
                X[col].fillna(X[col].mode()[0], inplace=True)
        print("✅ Missing values handled")
    else:
        print("✅ No missing values found")
    
    # Encode categorical variables
    categorical_columns = X.select_dtypes(include=['object']).columns
    numerical_columns = X.select_dtypes(include=['int64', 'float64']).columns
    
    print(f"\nCategorical columns: {len(categorical_columns)}")
    print(f"Numerical columns: {len(numerical_columns)}")
    
    # Label encode categorical variables
    label_encoders = {}
    for col in categorical_columns:
        le = LabelEncoder()
        X[col] = le.fit_transform(X[col].astype(str))
        label_encoders[col] = le
        print(f"  Encoded: {col} ({X[col].nunique()} unique values)")
    
    # Scale numerical features
    scaler = StandardScaler()
    if len(numerical_columns) > 0:
        X[numerical_columns] = scaler.fit_transform(X[numerical_columns])
        print(f"✅ Scaled {len(numerical_columns)} numerical columns")
    
    print(f"\nFinal feature matrix shape: {X.shape}")
    return X, label_encoders, scaler

# Preview the preprocessing on our 24 columns
print(f"\n🔍 PREPROCESSING PREVIEW:")
print("-" * 50)
X_processed, encoders, scaler = preprocess_features(df, columns_to_check)

print(f"\n📊 FEATURE STATISTICS:")
print("-" * 30)
print(X_processed.describe())
```

```python
# Model 1: Logistic Regression (Sigmoid) for Classification
print("🔵 MODEL 1: LOGISTIC REGRESSION (SIGMOID)")
print("=" * 80)

# Split data for classification
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_processed, y_binary, test_size=0.2, random_state=42, stratify=y_binary
)

print(f"Training set: {X_train_clf.shape[0]} samples")
print(f"Test set: {X_test_clf.shape[0]} samples")
print(f"Features: {X_train_clf.shape[1]}")

# Train Logistic Regression
print(f"\n🔄 Training Logistic Regression...")
logistic_model = LogisticRegression(random_state=42, max_iter=1000)
logistic_model.fit(X_train_clf, y_train_clf)

# Make predictions
y_pred_logistic = logistic_model.predict(X_test_clf)
y_prob_logistic = logistic_model.predict_proba(X_test_clf)[:, 1]

# Evaluate model
accuracy_logistic = accuracy_score(y_test_clf, y_pred_logistic)
print(f"✅ Logistic Regression trained successfully!")
print(f"\n📊 LOGISTIC REGRESSION RESULTS:")
print(f"   Accuracy: {accuracy_logistic:.4f}")

print(f"\n📈 Classification Report:")
print(classification_report(y_test_clf, y_pred_logistic))

print(f"\n🎯 Confusion Matrix:")
cm_logistic = confusion_matrix(y_test_clf, y_pred_logistic)
print(cm_logistic)

# Feature importance (coefficients)
feature_importance_logistic = pd.DataFrame({
    'feature': columns_to_check,
    'coefficient': logistic_model.coef_[0],
    'abs_coefficient': np.abs(logistic_model.coef_[0])
}).sort_values('abs_coefficient', ascending=False)

print(f"\n🔍 TOP 10 MOST IMPORTANT FEATURES (by coefficient magnitude):")
print("-" * 60)
for i, row in feature_importance_logistic.head(10).iterrows():
    print(f"{row.name+1:2d}. {row['coefficient']:8.4f} | {row['feature']}")

# Store results
logistic_results = {
    'model': logistic_model,
    'accuracy': accuracy_logistic,
    'predictions': y_pred_logistic,
    'probabilities': y_prob_logistic,
    'feature_importance': feature_importance_logistic
}
```

```python
# Model 2: Linear Regression for Continuous Target
print("🟢 MODEL 2: LINEAR REGRESSION")
print("=" * 80)

# Split data for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_processed, y_continuous, test_size=0.2, random_state=42
)

print(f"Training set: {X_train_reg.shape[0]} samples")
print(f"Test set: {X_test_reg.shape[0]} samples")
print(f"Features: {X_train_reg.shape[1]}")

# Train Linear Regression
print(f"\n🔄 Training Linear Regression...")
linear_model = LinearRegression()
linear_model.fit(X_train_reg, y_train_reg)

# Make predictions
y_pred_linear = linear_model.predict(X_test_reg)

# Evaluate model
mse_linear = mean_squared_error(y_test_reg, y_pred_linear)
rmse_linear = np.sqrt(mse_linear)
r2_linear = r2_score(y_test_reg, y_pred_linear)

print(f"✅ Linear Regression trained successfully!")
print(f"\n📊 LINEAR REGRESSION RESULTS:")
print(f"   Mean Squared Error (MSE): {mse_linear:.4f}")
print(f"   Root Mean Squared Error (RMSE): {rmse_linear:.4f}")
print(f"   R² Score: {r2_linear:.4f}")
print(f"   Mean Absolute Error: {np.mean(np.abs(y_test_reg - y_pred_linear)):.4f}")

# Feature importance (coefficients)
feature_importance_linear = pd.DataFrame({
    'feature': columns_to_check,
    'coefficient': linear_model.coef_,
    'abs_coefficient': np.abs(linear_model.coef_)
}).sort_values('abs_coefficient', ascending=False)

print(f"\n🔍 TOP 10 MOST IMPORTANT FEATURES (by coefficient magnitude):")
print("-" * 60)
for i, row in feature_importance_linear.head(10).iterrows():
    print(f"{row.name+1:2d}. {row['coefficient']:8.4f} | {row['feature']}")

# Create a simple residuals plot
print(f"\n📈 PREDICTION vs ACTUAL SUMMARY:")
print(f"   Actual range: [{np.min(y_test_reg):.3f}, {np.max(y_test_reg):.3f}]")
print(f"   Predicted range: [{np.min(y_pred_linear):.3f}, {np.max(y_pred_linear):.3f}]")
print(f"   Residuals mean: {np.mean(y_test_reg - y_pred_linear):.6f}")
print(f"   Residuals std: {np.std(y_test_reg - y_pred_linear):.4f}")

# Store results
linear_results = {
    'model': linear_model,
    'mse': mse_linear,
    'rmse': rmse_linear,
    'r2': r2_linear,
    'predictions': y_pred_linear,
    'feature_importance': feature_importance_linear
}
```
